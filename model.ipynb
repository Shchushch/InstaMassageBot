```python
def decoder(dic):
    for k,v in dic.items():
        #print(type(v))
        if type(v) == type([]):
            for el in v:
                if type(el) == type({'key':'value'}):
                    el.update(decoder(el))
        elif type(v) == type({'key':'value'}):
            v.update(decoder(v))
        elif type(v) == type('str'):
            try:
                dic.update({k:v.encode("latin1").decode("utf-8")})
            except:
                continue

    return dic
def dialoger(dialog):
    dialog=dialog['messages']
    for i in range(len(dialog)):
        dialog[i] = decoder(dialog[i])
    dialog= dialog[::-1]
    train='‚Äî ' + dialog[0]['content']
    for i in range(1,len(dialog)):
        
        if train[-1] not in '.!?)(':
            train+='.'
        train+=' '    
        if dialog[i]['sender_name']!=dialog[i-1]['sender_name']:
            train+='\n\n‚Äî '
        train+=dialog[i]['content']
    return train
```
